```python
# Оставим в отзывах только слова, а также переведем их в нижний регистр
def just_words_lower_case(text):
    text = re.sub('[^А-Яа-я]+', ' ', text).lower()
    return text

df['Review'] = df['Review'].apply(lambda x: just_words_lower_case(x))
df.head(1)
```

```python
segmenter = Segmenter()
morph_vocab = MorphVocab()

emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
syntax_parser = NewsSyntaxParser(emb)
ner_tagger = NewsNERTagger(emb)

def natasha_lemmatize(text):
    doc = Doc(text)
    doc.segment(segmenter)
    doc.tag_morph(morph_tagger)
    doc.parse_syntax(syntax_parser)
    doc.tag_ner(ner_tagger)
    
    result = []
    for token in doc.tokens:
        token.lemmatize(morph_vocab)
        result.append(token.lemma)
    
    return ' '.join(result)

df['Review'] = df['Review'].apply(lambda x: natasha_lemmatize(x))
df.head(1)
```
